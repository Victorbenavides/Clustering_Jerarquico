```python
import numpy as np #Para los arrays y calculos
import pandas as pd #analisis y carga de datos o archivos
import matplotlib.pyplot as plt #para graficar
import scipy.cluster.hierarchy as sch #para aplicar el clustering y la jerarquia
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
ciudades = pd.read_csv('ClimaCoahuila.csv', engine='python', index_col=0) #se cargan los datos desde un archivo csv
```

```python
ciudades.info()#se muestran las columnas
```

```python
ciudades.head()#se muestra solo la cabecera
```

```python
clima = ciudades.iloc[:, [0,1,2,3,4,5,6]].values #con iloc seleccionamos los datos de las columnas
clima#se imprimen los datos
```

```python
CJerarquico = linkage(clima, 'single')#Se aplica el clustering jerarquico
```

```python
dendrogram=sch.dendrogram(CJerarquico)#Se muestra el dendrograma
```

```python
clusters = fcluster(CJerarquico, t=5, criterion='distance') #Se crean los clusters, con t es la altura del corte y pedimos la distancia maxima
```

```python
clusters #Se imprimen los clusters
```

```python
ciudades['Clustering Jerarquico'] = clusters #Se pasan los clusters a la variable ciudades
ciudades #Se imprimen los clusters
```
